# Solutions 2

```python
import matplotlib.pyplot as plt
```

```python
import skimage.data
```

```python
dir(skimage.data)
```

```python
image = skimage.data.moon()
```

```python
plt.imshow(image)
plt.show()
```

```python
plt.imshow(image, cmap = 'jet')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image, cmap = 'gray')
plt.imshow(image, cmap = 'jet', alpha = 0.3)
plt.show()
```
